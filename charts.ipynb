```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

RESULTS_DIR = "results/basic_tests"
FILE_NAME = "generations.csv"
```

```python
task_dir = "1_1_A"

FILE_PATH = os.path.join(RESULTS_DIR, task_dir, FILE_NAME)

df = pd.read_csv(FILE_PATH, sep=",")

generations = df["Generation"]
best_fitness = df["Best Fitness"]
average_fitness = df["Average Fitness"]

plt.scatter(generations, best_fitness, label="Best Fitness")
plt.scatter(generations, average_fitness, label="Average Fitness")
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Fitness")

plt.show()
```

```python
task_dir = "1_1_B"

FILE_PATH = os.path.join(RESULTS_DIR, task_dir, FILE_NAME)

df = pd.read_csv(FILE_PATH, sep=",")

generations = df["Generation"]
best_fitness = df["Best Fitness"]
average_fitness = df["Average Fitness"]

fig, axs = plt.subplots(2, 1)

axs[0].plot(generations, best_fitness)
axs[0].set_ylabel("Najlepsze przystosowanie")

axs[1].plot(generations, average_fitness, color="orange")
axs[1].set_ylabel("rednie przystosowanie")

plt.xlabel("Generacja")

plt.show()
```
